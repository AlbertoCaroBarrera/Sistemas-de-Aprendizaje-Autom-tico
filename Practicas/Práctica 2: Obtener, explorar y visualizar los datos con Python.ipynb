# Importar Librerías Necesarias
Importa las librerías necesarias, incluyendo pandas y matplotlib.

```python
# Importar Librerías Necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

# Cargar Datos desde CSV
Carga los datos desde el archivo CSV proporcionado usando pandas.

```python
# Cargar Datos desde CSV
data_url = "https://raw.githubusercontent.com/ageron/data/main/lifesat/lifesat.csv"
data = pd.read_csv(data_url)

# Mostrar las primeras filas del DataFrame para verificar la carga correcta
data.head()
```

# Explorar Datos
Realiza una exploración superficial del conjunto de datos, incluyendo la visualización de las primeras filas y la obtención de estadísticas descriptivas.

```python
# Explorar Datos

# Mostrar las primeras filas del DataFrame para verificar la carga correcta
data.head()

# Obtener estadísticas descriptivas del DataFrame
data.describe()

# Visualizar la distribución de los datos
data.hist(bins=50, figsize=(20, 15))
plt.show()
```

# Visualizar Datos
Crea gráficos para visualizar los datos, como gráficos de dispersión y diagramas de caja.

```python
# Visualizar Datos

# Crear un gráfico de dispersión para visualizar la relación entre el PIB per cápita y la satisfacción con la vida
data.plot(kind="scatter", x="GDP per capita (USD)", y="Life satisfaction", alpha=0.5)
plt.title("Relación entre PIB per cápita y Satisfacción con la Vida")
plt.xlabel("PIB per cápita")
plt.ylabel("Satisfacción con la Vida")
plt.grid(True)
plt.show()

# Crear diagramas de caja para visualizar la distribución de la satisfacción con la vida
data.boxplot(column="Life satisfaction", vert=False)
plt.title("Diagrama de Caja de la Satisfacción con la Vida")
plt.xlabel("Satisfacción con la Vida")
plt.show()
```
